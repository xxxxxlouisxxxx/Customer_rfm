```python
for feature in ["Quantity","Price","TotalPrice"]:

    Q1 = df[feature].quantile(0.01)
    Q3 = df[feature].quantile(0.99)
    IQR = Q3-Q1
    upper = Q3 + 1.5*IQR
    lower = Q1 - 1.5*IQR

    if df[(df[feature] > upper) | (df[feature] < lower)].any(axis=None):
        print(feature,"yes")
        print(df[(df[feature] > upper) | (df[feature] < lower)].shape[0])
    else:
        print(feature, "no")
```

```python
df2['InvoiceDate_parsed'] = pd.to_datetime(
    df2['InvoiceDate'],
    dayfirst=True,
    infer_datetime_format=True,
    errors='coerce'
)
```

```python
df2.groupby("Customer ID").agg({"InvoiceDate":"max"}).head()
```

For each customer, we need to subtract the customers' last purchase date from today's date.

```python
df2['InvoiceDate'] = df2['InvoiceDate_parsed']

today = pd.Timestamp(today_date)

# Calculate recency (days since last purchase)
last_purchase = df2.groupby('Customer ID')['InvoiceDate'].max()
recency = (today - last_purchase).dt.days

# Keep as a temp dataframe
recency_df = recency.reset_index()
recency_df.columns = ['Customer ID', 'InvoiceDate']
recency_df.head()
```

```python
temp_df = (today_date - df2.groupby("Customer ID").agg({"InvoiceDate":"max"}))
```

Create a new data frame --> rfm

```python
rfm = pd.concat([recency_df, freq_df, monetary_df],  axis=1)
rfm.head()
```

1 as the worst recency value ; 5 as the best recency value

the smaller recency value means more recent buy

```python
rfm["RecencyScore"] = pd.qcut(rfm['Recency'], 5, labels = [5, 4, 3, 2, 1])
```

```python
rfm
```

```python
rfm["Frequency"].describe([0.01,0.05,0.10,0.25,0.50,0.60,0.75,0.90,0.95, 0.99]).T
```

```python
cut_bins = [0,1,2,3,9,210]
```

```python
rfm["FrequencyScore"] = pd.cut(rfm["Frequency"], bins = cut_bins, labels = [1, 2, 3, 4, 5])
rfm["FrequencyScore"].value_counts()
```

```python
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 5, labels = [1, 2, 3, 4, 5])
rfm.head()
```

```python
(rfm['RecencyScore'].astype(str) + 
 rfm['FrequencyScore'].astype(str) + 
 rfm['MonetaryScore'].astype(str)).head()
```

```python
rfm["RFM_SCORE"] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str) + rfm['MonetaryScore'].astype(str)
rfm.head()
```

```python
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Lose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
#Pull the value corresponding to the ranges of the score in the dictionary defined in 'seg_map', 
#and add it as a new column named "Segment":

rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
rfm['Segment'].head()
```

```python
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm.head()
```

```python
need_attention_df = pd.DataFrame()
need_attention_df["Need_Attention_CustomerID"] = rfm[rfm["Segment"] == "Need Attention"].index
need_attention_df.head()
```
